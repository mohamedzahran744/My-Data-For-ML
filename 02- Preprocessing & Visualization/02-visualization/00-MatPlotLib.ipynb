## Scatter Plot

Scatter plots are used to visualize the relationship between two numerical variables. Each point represents an observation, with its position determined by the values of the two variables on the x and y axes.

**Use Cases:**
- Identifying correlations between variables
- Detecting patterns, trends, or outliers
- Comparing two continuous variables

---

```python
# Create a scatter plot to visualize the relationship between longitude and latitude
# plt.scatter() plots individual data points as dots
# x-axis: longitude values from the DataFrame
# y-axis: latitude values from the DataFrame
plt.scatter(df['longitude'], df['latitude'])

# Label the x-axis to indicate it represents longitude
plt.xlabel('longitude')

# Label the y-axis to indicate it represents latitude
plt.ylabel('latitude')

# Add a descriptive title to the plot for context
plt.title('Longitude vs. Latitude of Housing Dataset')

# Display the complete plot in the notebook output
# Note: plt.show() is optional when using %matplotlib inline, but good practice
plt.show()
```

```python
# Advanced scatter plot with extensive customization options
# Demonstrating how to control appearance, size, colors, and styling

# Create a new figure with custom dimensions
# figsize=(width, height) in inches - makes the plot larger and more readable
plt.figure(figsize=(10, 6))

# Create scatter plot with customization parameters:
# x, y: data columns for x and y axes
# marker='o': use circular markers for data points
# color='black': set all points to black color
# cmap: colormap option (jet) - not visible here since color='black' overrides it
plt.scatter(x=df['longitude'], y=df['latitude'], marker='o', color='black', 
            cmap=plt.get_cmap('jet'))

# Customize x-axis label with larger font size and red color
plt.xlabel('Longitude', fontsize=14, color='red')

# Customize y-axis label with larger font size and red color
plt.ylabel('Latitude', fontsize=14, color='red')

# Customize plot title with specific styling
# label: the title text
# fontsize: makes title larger and more prominent
# color: red color for emphasis
plt.title(label='Longitude vs. Latitude of Housing Dataset', fontsize=15, color='red')

# Display the fully customized plot
plt.show()
```

## Line Plot

Line plots are used to visualize trends and changes over a continuous interval or time period. They connect data points with lines, making it easy to see patterns, trends, and fluctuations.

**Use Cases:**
- Tracking changes over time (time series data)
- Showing trends and patterns in sequential data
- Comparing multiple series on the same axes
- Visualizing continuous relationships between variables

---

```python
# Create test datasets for demonstrating line plots with different mathematical functions

# Generate 100 evenly spaced values between 1 and 5
# np.linspace(start, stop, num) creates a linear sequence
# This will be used as the x-axis values for our plots
x = np.linspace(1, 5, 100) 

# Calculate exponential values: e^x for each x value
# Exponential function shows rapid growth - useful for demonstrating curved trends
y = np.exp(x)  

# Calculate logarithm base 2 values: log2(x) for each x value
# Logarithmic function shows diminishing growth rate - opposite of exponential
z = np.log2(x)
```

```python
# Create a simple line plot showing exponential growth
# plt.plot() connects data points with lines (default behavior)
# x-axis: linear values from 1 to 5
# y-axis: exponential values (e^x)
plt.plot(x, y)

# Add a descriptive title to identify what variables are being plotted
plt.title('x vs. y')

# Label the x-axis to indicate the independent variable
plt.xlabel('x')

# Label the y-axis to indicate the dependent variable (exponential values)
plt.ylabel('y')

# Display the line plot showing the exponential curve
plt.show()
```

```python
## twinx

## We want to optimize the above figure
fig, ax = plt.subplots()

ax.plot(x, y, color='blue', label='exp(x)')
ax.set_xlabel('x', fontsize=14)
ax.set_ylabel('y', color='blue', fontsize=14)

## create another axis
ax2 = ax.twinx()
ax2.plot(x, z, color='red', label='log10(x)')

ax2.set_ylabel('z', color='red', fontsize=14)

plt.title('x vs. y & x vs. z', fontsize=14)
plt.show()
```

```python
# Using twinx() to create dual y-axes for better visualization of different scales
# This is essential when comparing functions with vastly different value ranges

# Create a figure and primary axis object using subplots
# This gives us more control than basic plt.plot()
fig, ax = plt.subplots()

# Plot exponential function on the primary (left) y-axis
# Blue color to match the left y-axis
ax.plot(x, y, color='blue', label='exp(x)')

# Label the x-axis (shared by both functions)
ax.set_xlabel('x', fontsize=14)

# Label the left y-axis for exponential values
# Color matches the line for clarity
ax.set_ylabel('y', color='blue', fontsize=14)

# Create a second y-axis sharing the same x-axis
# twinx() creates a twin axis on the right side
# This allows plotting data with different scales without distortion
ax2 = ax.twinx()

# Plot logarithmic function on the secondary (right) y-axis
# Red color to match the right y-axis
ax2.plot(x, z, color='red', label='log10(x)')

# Label the right y-axis for logarithmic values
# Color matches the line for easy identification
ax2.set_ylabel('z', color='red', fontsize=14)

# Add title describing both relationships
plt.title('x vs. y & x vs. z', fontsize=14)

# Display the plot with dual y-axes
plt.show()
```

```python
# Adding annotations to highlight specific points or features in the plot
# Annotations are useful for drawing attention to important data points or trends

# Create a line plot of the exponential function
plt.plot(x, y)

# Add title to describe the plot
plt.title('x vs. y')

# Label the x-axis
plt.xlabel('x')

# Label the y-axis  
plt.ylabel('y')

# Add an annotation with an arrow pointing to a specific location
# 'max value': text label for the annotation
# xy=(5, 140): coordinates of the point being annotated (where arrow points)
# xytext=(4, 20): coordinates where the text label is placed
# arrowprops: dictionary defining arrow appearance
#   - arrowstyle="->": simple arrow with pointer
#   - color="red": red arrow for visibility
plt.annotate('max value', xy=(5, 140), xytext=(4, 20), 
             arrowprops={"arrowstyle":"->","color":"red"})

# Display the annotated plot
plt.show()
```

## Histogram

Histograms display the distribution of a numerical variable by dividing the data into bins and showing the frequency of observations in each bin. They reveal the shape, central tendency, and spread of the data.

**Use Cases:**
- Understanding data distribution (normal, skewed, bimodal, etc.)
- Identifying outliers and data ranges
- Comparing distributions across different groups
- Analyzing frequency and density of values

---

```python
# Create a histogram to visualize the distribution of housing ages
# Histograms show how frequently different age values occur in the dataset

# Set figure size for better readability
plt.figure(figsize=(10, 6))

# Create histogram of housing median age column
# bins=50: divide the data range into 50 intervals (bars)
# More bins = more detail but potentially noisier
# Fewer bins = smoother but less detail
plt.hist(df['housing_median_age'], bins=50)

# Add descriptive title to explain what distribution is shown
plt.title('Histogram Distr. for Houses ages', fontsize=12)

# Label x-axis to indicate what variable is being measured
plt.xlabel('Housing median age', fontsize=12)

# Label y-axis to show frequency (count of houses in each bin)
plt.ylabel('Frequency', fontsize=12)

# Display the histogram showing age distribution pattern
plt.show()
```

```python
# Create a normalized histogram with outline style
# This variation shows probability density rather than raw counts

# Set figure size for better readability
plt.figure(figsize=(10, 6))

# Create histogram with normalization and outline styling:
# bins=50: divide data into 50 intervals
# density=True: normalize the histogram so the area under the curve equals 1
#               Y-axis shows probability density instead of raw frequency counts
#               Useful for comparing distributions of different sample sizes
# histtype='step': display as an unfilled outline instead of filled bars
#                  Creates a cleaner, line-based visualization
plt.hist(df['housing_median_age'], bins=50, density=True, histtype='step')

# Add descriptive title
plt.title('Histogram Distr. for Houses ages', fontsize=12)

# Label x-axis showing the variable being analyzed
plt.xlabel('Housing median age', fontsize=12)

# Label y-axis - note: with density=True, this is now probability density, not raw frequency
plt.ylabel('Frequency', fontsize=12)

# Display the normalized outline histogram
plt.show()
```

## Box Plot

Box plots (also called box-and-whisker plots) provide a visual summary of the distribution of numerical data through quartiles. They show the median, quartiles, and potential outliers in a compact format.

**Key Components:**
- **Box**: Represents the interquartile range (IQR) - middle 50% of data
- **Line inside box**: Median (50th percentile)
- **Whiskers**: Extend to show the range of most data points
- **Points beyond whiskers**: Potential outliers

**Use Cases:**
- Identifying outliers and extreme values
- Comparing distributions across multiple groups
- Understanding data spread and skewness
- Summarizing key statistical measures visually

---

```python
# Create a box plot to visualize the distribution and outliers of median income
# Box plots show median, quartiles, range, and outliers in a compact visualization

# Set figure size for better readability
plt.figure(figsize=(10, 6))

# Create box plot for the median_income column
# Shows: minimum, Q1 (25th percentile), median (Q2/50th percentile), 
#        Q3 (75th percentile), maximum, and outliers
plt.boxplot(df['median_income'])  

# Alternative with explicit whisker parameter (currently commented out):
# whis=1.5: whiskers extend to 1.5 * IQR beyond Q1 and Q3 (default value)
#           Points beyond whiskers are marked as outliers
# plt.boxplot(df['median_income'], whis=1.5)

# Add descriptive title explaining which variable is visualized
plt.title('Boxplot for median_income variable', fontsize=12)

# Label x-axis to identify the variable
plt.xlabel('median_income', fontsize=12)

# Label y-axis to indicate these are the actual income values
plt.ylabel('values', fontsize=12)

# Display the box plot showing income distribution and any outliers
plt.show()
```

## Bar Plot

Bar plots display categorical data with rectangular bars, where the length of each bar is proportional to the value it represents. They are ideal for comparing values across different categories.

**Key Features:**
- Bars can be vertical (bar plot) or horizontal (barh plot)
- Height/length represents the magnitude of each category
- Easy comparison between discrete categories

**Use Cases:**
- Comparing quantities across different categories
- Showing counts or frequencies of categorical variables
- Displaying aggregated metrics (sum, mean, count) by group
- Ranking items by value

---

```python
# Create a bar plot showing the frequency distribution of ocean proximity categories
# Using pandas' built-in plotting for concise syntax

# df['ocean_proximity'].value_counts(): counts occurrences of each unique category
#                                       returns a Series with categories as index and counts as values
# .plot(kind='bar'): creates a vertical bar chart directly from the value counts
#                    each bar represents one category, height shows frequency
df['ocean_proximity'].value_counts().plot(kind='bar')

# Display the bar plot showing which ocean proximity categories are most/least common
plt.show()
```

```python
# Create a bar plot using matplotlib for more control over customization
# This approach separates data extraction from plotting for flexibility

# Extract the count values (frequencies) for each ocean proximity category
# .values returns a numpy array of the counts
vals = df['ocean_proximity'].value_counts().values

# Extract the category names (ocean proximity types) as the bar labels
# .index returns the unique category names
idx = df['ocean_proximity'].value_counts().index

# Set figure size for better readability and spacing
plt.figure(figsize=(10, 6))

# Create vertical bar plot:
# x=idx: positions/labels for bars (category names)
# height=vals: height of each bar (frequency counts)
plt.bar(x=idx, height=vals)

# Add descriptive title explaining the visualization
plt.title('Distribution for ocean_proximity in Dataset', fontsize=14)

# Label x-axis to identify what categories represent
plt.xlabel('Ocean Proximity', fontsize=12)

# Label y-axis to show these are frequency counts
plt.ylabel('Counts', fontsize=12)

# Rotate x-axis labels 90 degrees to prevent overlapping text
# Essential when category names are long or numerous
plt.xticks(rotation=90)

# Display the bar plot with all customizations applied
plt.show()
```

## Pie Chart

Pie charts display data as slices of a circle, where each slice represents a proportion of the whole. The size of each slice is proportional to its percentage of the total.

**Key Features:**
- Shows parts of a whole (proportions/percentages)
- Circle represents 100% of the data
- Each slice angle is proportional to its value

**Use Cases:**
- Showing percentage breakdowns of categorical data
- Visualizing composition or market share
- Displaying budget allocations or resource distribution
- Comparing relative proportions (best with 5-7 categories or fewer)

**Note:** Pie charts work best with a small number of categories. For many categories, bar charts are often more readable.

---

```python
# Create an enhanced pie chart showing the distribution of ocean proximity categories
# Demonstrates advanced customization options for visual impact

# df['ocean_proximity'].value_counts(): get frequency of each category
# .plot(kind='pie'): create a pie chart from the value counts
# figsize=(10, 6): set figure dimensions for better visibility
# explode=(0.1, 0.2, 0.05, 0.4, 0.2): separate slices from center
#         Each value corresponds to a slice; higher values = more separation
#         Creates visual emphasis on specific categories
# autopct='%1.2f%%': display percentages on each slice
#         %1.2f = floating point with 2 decimal places, %% = percent symbol
# shadow=True: add shadow effect for 3D appearance and depth
# colors=(...): custom color palette for each slice
#         Tuple of colors matching the number of categories
df['ocean_proximity'].value_counts().plot(kind='pie', figsize=(10, 6), explode=(0.1, 0.2, 0.05, 0.4, 0.2), autopct='%1.2f%%', 
        shadow=True, colors=('orange', 'cyan', 'brown', 'grey', 'indigo'))

# Display the customized pie chart showing proportional distribution
plt.show()
```

## Subplots

**What about subplots?**

Subplots allow you to display multiple plots in a single figure, arranged in a grid layout. This is perfect for comparing related visualizations side-by-side or showing different aspects of your data together.

**Use Case Example:**  
Let's say we want to get the histogram for `median_house_value` at each `ocean_proximity` category separately. Instead of creating 5 separate figures, we can use subplots to display all distributions in one organized view for easy comparison.

**Benefits:**
- Compare multiple related plots simultaneously
- Save space and create compact visualizations
- Identify patterns across different subsets of data
- Create publication-ready multi-panel figures

---

```python
## check
df['ocean_proximity'].value_counts()
```

```python
## example
df[df['ocean_proximity']=='INLAND'].head()
```

```python
# Create a 2x2 grid of subplots to compare house price distributions across ocean proximity categories
# Subplots enable side-by-side comparison of distributions for different data subsets

# Create figure with 2 rows and 2 columns of subplots
# fig: the overall figure object
# ax: 2D array of axes objects for accessing individual subplots (ax[row, col])
# figsize=(20, 10): large figure to accommodate 4 histograms clearly
fig, ax = plt.subplots(2, 2, figsize=(20, 10))

# Add padding between subplots to prevent overlapping titles and labels
# pad=5: spacing in font-size units between subplots
fig.tight_layout(pad=5)

# TOP LEFT subplot (row 0, column 0): Houses less than 1 hour from ocean
plt.sca(ax[0, 0])  # Set current axes to top-left subplot
# Filter dataframe for houses <1H from ocean and extract median house values
to_plot = df[df['ocean_proximity']=='<1H OCEAN']['median_house_value']
plt.hist(to_plot, bins=50)  # Create histogram with 50 bins
plt.title('Distr. of Houses Prices at ocean_proximity=<1H OCEAN', fontsize=14)
plt.ylabel('Frequency', fontsize=14)  # Y-axis label (left side subplots only)

# TOP RIGHT subplot (row 0, column 1): Inland houses
plt.sca(ax[0, 1])  # Set current axes to top-right subplot
# Filter for inland houses
to_plot = df[df['ocean_proximity']=='INLAND']['median_house_value']
plt.hist(to_plot, bins=50)
plt.title('Distr. of Houses Prices at ocean_proximity=INLAND', fontsize=14)

# BOTTOM LEFT subplot (row 1, column 0): Houses near ocean
plt.sca(ax[1, 0])  # Set current axes to bottom-left subplot
# Filter for houses near ocean
to_plot = df[df['ocean_proximity']=='NEAR OCEAN']['median_house_value']
plt.hist(to_plot, bins=50)
plt.title('Distr. of Houses Prices at ocean_proximity=NEAR OCEAN', fontsize=14)
plt.xlabel('Houses Prices', fontsize=14)  # X-axis label (bottom row subplots)
plt.ylabel('Frequency', fontsize=14)

# BOTTOM RIGHT subplot (row 1, column 1): Houses near bay
plt.sca(ax[1, 1])  # Set current axes to bottom-right subplot
# Filter for houses near bay
to_plot = df[df['ocean_proximity']=='NEAR BAY']['median_house_value']
plt.hist(to_plot, bins=50)
plt.title('Distr. of Houses Prices at ocean_proximity=NEAR BAY', fontsize=14)
plt.xlabel('Houses Prices', fontsize=14)

# Display the complete 2x2 grid showing price distribution comparisons
plt.show()
```

```python
# Optimized approach: Use loops to create subplots programmatically
# This reduces code repetition and makes it easier to maintain and scale

# Create a 2x2 grid of subplots
fig, ax = plt.subplots(2, 2, figsize=(20, 10))

# Add spacing between subplots to prevent overlap
fig.tight_layout(pad=5)

# Get the top 4 most common ocean_proximity categories
# .value_counts() sorts categories by frequency (descending)
# .index[:4] selects the first 4 category names
# .tolist() converts to a Python list
# This automatically excludes the 5th category which has very few observations
ocean_vals = df['ocean_proximity'].value_counts().index[:4].tolist()

# Reshape the 1D list into a 2x2 array to match subplot grid structure
# This allows us to access categories using [row, col] indexing
ocean_vals = np.array(ocean_vals).reshape(2, 2)

# Nested loops iterate through each subplot position
for row in range(2):  # Loop through rows (0, 1)
    for col in range(2):  # Loop through columns (0, 1)
        
        # Set the current active subplot to ax[row, col]
        plt.sca(ax[row, col])
        
        # Filter data for the current ocean proximity category
        # ocean_vals[row, col] gets the category name for this subplot position
        to_plot = df[df['ocean_proximity']==ocean_vals[row, col]]['median_house_value']
        
        # Create histogram for the filtered data
        plt.hist(to_plot, bins=50)
        
        # Dynamic title using f-string to insert the category name
        plt.title(f'Distr. of Houses Prices at ocean_proximity={ocean_vals[row, col]}', fontsize=14)
        
        # Conditional label placement for clean appearance:
        # X-axis labels only on bottom row (row 1)
        if row != 0:
            plt.xlabel('Houses Prices', fontsize=14)
        
        # Y-axis labels only on left column (col 0)
        if col != 1:
            plt.ylabel('Frequency', fontsize=14)

# Display the programmatically generated subplot grid
plt.show()
```
