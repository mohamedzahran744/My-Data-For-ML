# Seaborn - Statistical Data Visualization

## What is Seaborn?

Seaborn is a Python data visualization library built on top of Matplotlib. It provides a high-level interface for creating attractive and informative statistical graphics with less code.

## Key Advantages:
✨ **Beautiful default styles** - Professional-looking plots out of the box  
✨ **Statistical functions** - Built-in support for statistical estimation and visualization  
✨ **Complex plots made easy** - Simplified syntax for advanced visualizations  
✨ **Color palettes** - Sophisticated color schemes and themes  
✨ **Integration** - Works seamlessly with pandas DataFrames  

## Common Use Cases:
- Distribution plots (distplot, histplot, kdeplot)
- Categorical plots (boxplot, violinplot, swarmplot)
- Relationship plots (scatterplot, lineplot, heatmap)
- Matrix plots (correlation heatmaps, clustermaps)
- Regression plots with confidence intervals

Let's explore Seaborn's powerful visualization capabilities!

## Count Plot - Categorical Bar Plot

The **count plot** is Seaborn's version of a bar chart for categorical data. It automatically counts the occurrences of each category and displays them as bars.

**Key Features:**
- Automatically counts frequency of each category (no manual counting needed)
- Cleaner syntax than matplotlib's bar plot
- Built-in statistical estimation
- Easy grouping and color coding by additional variables

**Comparison to Matplotlib:**
- Matplotlib: Requires manual `value_counts()` before plotting
- Seaborn: Automatically computes counts from raw data

**Use Cases:**
- Visualizing distribution of categorical variables
- Comparing frequencies across different groups
- Quick exploratory data analysis of categories

---

```python
# Create a count plot to visualize the distribution of meal times (Lunch vs Dinner)
# Seaborn automatically counts occurrences without needing value_counts()

# Set figure size for better visibility
plt.figure(figsize=(10, 8))

# Create count plot with color coding:
# x='time': categorical variable to plot on x-axis (values to count)
# data=df_tips: specify the DataFrame containing the data
# hue='time': color-code bars by the same variable (creates legend)
#             Using hue with the same variable as x adds color distinction
sns.countplot(x='time', data=df_tips, hue='time')

# Add descriptive title explaining what distribution is shown
plt.title('Distr. of time variable in Dataset', fontsize=14)

# Label x-axis to identify the categorical variable
plt.xlabel('Time', fontsize=14)

# Label y-axis to indicate these are frequency counts
plt.ylabel('Count', fontsize=14)

# Display the count plot showing meal time distribution
plt.show()
```

```python
# Create a grouped count plot to compare meal times between male and female customers
# This shows how an additional categorical variable (sex) is distributed within each time category

# Set figure size for clear visualization
plt.figure(figsize=(10, 8))

# Create grouped count plot:
# x='time': main categorical variable (Lunch/Dinner) on x-axis
# data=df_tips: source DataFrame
# hue='sex': group and color-code bars by sex (Male/Female)
#            Creates side-by-side bars for each time category
#            Allows comparison of gender distribution at lunch vs dinner
sns.countplot(x='time', data=df_tips, hue='sex')

# Add title describing the grouped comparison
plt.title('Distr. of time variable in Dataset', fontsize=12)

# Label x-axis for the main categorical variable
plt.xlabel('Time', fontsize=12)

# Label y-axis to show frequency counts
plt.ylabel('Count', fontsize=12)

# Position legend in upper left corner to identify male vs female bars
plt.legend(loc='upper left')

# Display the grouped count plot
plt.show()
```

```python
# Alternative approach: Using catplot with kind='count' instead of countplot
# catplot is a figure-level function that provides more flexibility and automatic sizing

# NOTE: plt.figure() is not needed here because catplot creates its own figure
# This line will create a separate empty figure (can be removed)
plt.figure(figsize=(10, 8))

# Create count plot using the catplot interface:
# kind='count': specifies a count plot (equivalent to sns.countplot)
# x='time': categorical variable to count
# data=df_tips: source DataFrame
# hue='sex': group and color bars by gender
# catplot automatically handles figure creation and sizing
sns.catplot(kind='count', x='time', data=df_tips, hue='sex')

# Add descriptive title
# NOTE: With catplot, use plt.suptitle() or access the figure object for better control
plt.title('Distr. of time variable in Dataset', fontsize=12)

# Label axes
plt.xlabel('Time', fontsize=12)
plt.ylabel('Count', fontsize=12)

# Display the plot
plt.show()
```

## Box Plot in Seaborn

Seaborn's **box plot** provides an enhanced version of matplotlib's boxplot with better aesthetics and easier syntax for grouped comparisons.

**Key Features:**
- Automatic grouping by categorical variables
- Cleaner, more polished appearance
- Easy integration with pandas DataFrames
- Built-in support for multiple grouping variables

**Box Plot Components (Reminder):**
- **Box**: Interquartile range (IQR) - middle 50% of data
- **Line in box**: Median value
- **Whiskers**: Extend to 1.5 × IQR from quartiles
- **Points**: Outliers beyond the whiskers

**Advantages over Matplotlib:**
- No need to manually filter and plot each category
- Automatic handling of categorical grouping
- Beautiful default styling
- Support for nested grouping with `hue` parameter

**Use Cases:**
- Comparing distributions across categories
- Identifying outliers in different groups
- Visualizing statistical summaries by category

---

```python
# Create a box plot to visualize the distribution and outliers of total bill amounts
# Box plots are excellent for identifying the spread, median, and extreme values

# Set figure size for clear visualization
plt.figure(figsize=(10, 8))

# Create vertical box plot:
# y='total_bill': numerical variable to plot on y-axis
# data=df_tips: source DataFrame
# Creates a single box showing the distribution of all total bills
# Displays: median, Q1, Q3, whiskers (1.5×IQR), and outlier points
sns.boxplot(y='total_bill', data=df_tips)

# Add descriptive title
plt.title('Distr of total bill', fontsize=14)

# Label x-axis (though there's only one category here)
plt.xlabel('Total bill', fontsize=14)

# Label y-axis showing the actual bill values
plt.ylabel('Values', fontsize=14)

# Display the box plot showing bill distribution and any outliers
plt.show()
```

```python
# Create grouped box plots to compare total bill distributions between males and females
# This reveals if there are differences in spending patterns by gender

# NOTE: plt.figure() is redundant here since catplot creates its own figure
plt.figure(figsize=(10, 8))

# Create grouped box plots using catplot:
# y='total_bill': numerical variable (bill amounts) on y-axis
# x='sex': categorical variable (Male/Female) on x-axis for grouping
# data=df_tips: source DataFrame
# kind='box': specifies box plot type
# hue='sex': color-code boxes by gender (same as x variable here)
#            Adds color distinction to each gender's box
# catplot is a figure-level function that automatically handles sizing
sns.catplot(y='total_bill', x='sex', data=df_tips, kind='box', hue='sex')

# Add descriptive title
plt.title('Distr of total bill', fontsize=14)

# Label x-axis identifying the grouping variable
plt.xlabel('Sex', fontsize=14)

# Label y-axis showing the bill values
plt.ylabel('Values', fontsize=14)

# Display the comparative box plots
plt.show()
```

```python
# Compare total bill distributions between lunch and dinner times
# Box plots reveal if meal timing affects spending patterns

# NOTE: plt.figure() is not needed with catplot (figure-level function)
# catplot creates and manages its own figure automatically
plt.figure(figsize=(12, 8))

# Create grouped box plots comparing bills by meal time:
# y='total_bill': numerical variable (bill amounts) on y-axis
# x='time': categorical variable (Lunch/Dinner) for grouping
# data=df_tips: source DataFrame
# kind='box': specifies box plot visualization
# hue='time': color-code boxes by time (adds color distinction)
#             Using hue with same variable as x enhances visual separation
# Reveals differences in bill distributions between lunch and dinner
sns.catplot(y='total_bill', x='time', data=df_tips, kind='box', hue='time')

# Add descriptive title
plt.title('Distr of total bill', fontsize=14)

# Label x-axis showing the time categories
plt.xlabel('Time', fontsize=14)

# Label y-axis showing bill amount values
plt.ylabel('Values', fontsize=14)

# Display the comparison showing lunch vs dinner bill distributions
plt.show()
```

## Scatter Plot in Seaborn

Seaborn's **scatter plot** enhances matplotlib's scatter plots with better default aesthetics, easier grouping, and integrated statistical features.

**Key Features:**
- Clean, modern styling out of the box
- Automatic color coding by categorical variables
- Support for size variation based on data
- Easy addition of regression lines with `regplot` or `lmplot`
- Seamless integration with pandas DataFrames

**Advantages over Matplotlib:**
- Simpler syntax for grouped scatter plots
- Better default color palettes
- Built-in support for legends and grouping
- Automatic handling of categorical variables with `hue`

**Use Cases:**
- Exploring relationships between two numerical variables
- Identifying correlations and patterns
- Comparing relationships across different groups
- Adding regression lines to show trends

**Related Functions:**
- `scatterplot()`: Basic scatter plot
- `regplot()`: Scatter plot with regression line
- `lmplot()`: Figure-level scatter plot with regression

---

```python
# Create a scatter plot to explore the relationship between bill amount and tip
# This visualization helps identify if higher bills correlate with higher tips

# Set figure size for clear visualization
plt.figure(figsize=(10, 8))

# Create scatter plot:
# x='total_bill': independent variable (bill amount) on x-axis
# y='tip': dependent variable (tip amount) on y-axis
# data=df_tips: source DataFrame
# Each point represents one dining transaction
# Pattern reveals if tips increase proportionally with bill amount
sns.scatterplot(x='total_bill', y='tip', data=df_tips)

# Add descriptive title explaining the relationship being explored
plt.title('Relation between total bill and tip', fontsize=14)

# Label x-axis identifying the bill variable
plt.xlabel('Total bill', fontsize=14)

# Label y-axis identifying the tip variable
plt.ylabel('Tip', fontsize=14)

# Display the scatter plot showing bill-tip relationship
plt.show()
```

```python
# Enhanced scatter plot with color coding to compare tipping patterns by gender
# Adding the hue parameter reveals if males and females have different tipping behaviors

# Set figure size for clear visualization
plt.figure(figsize=(10, 8))

# Create color-coded scatter plot:
# x='total_bill': bill amount on x-axis
# y='tip': tip amount on y-axis
# data=df_tips: source DataFrame
# hue='sex': color-code points by gender (Male/Female)
#            Creates separate colors for each gender
#            Automatically generates a legend
#            Reveals if tipping patterns differ between genders
sns.scatterplot(x='total_bill', y='tip', data=df_tips, hue='sex')

# Add descriptive title
plt.title('Relation between total bill and tip', fontsize=14)

# Label x-axis
plt.xlabel('Total bill', fontsize=14)

# Label y-axis
plt.ylabel('Tip', fontsize=14)

# Display the color-coded scatter plot showing gender-based tipping patterns
plt.show()
```

## Subplots - Relational Plot (relplot)

Seaborn's **relplot** (relational plot) is a figure-level function that creates scatter or line plots with built-in support for creating subplot grids based on categorical variables.

**Key Features:**
- Automatically creates subplots using `col` and `row` parameters
- Faceting: splits data into multiple panels based on categories
- Consistent scaling across subplots for easy comparison
- Combines the power of scatter/line plots with automatic grid layout

**Parameters:**
- `kind='scatter'`: Creates scatter plots (default)
- `kind='line'`: Creates line plots
- `col='variable'`: Creates columns of subplots for each category
- `row='variable'`: Creates rows of subplots for each category
- `hue='variable'`: Color-codes within each subplot

**Advantages:**
- No manual subplot creation needed
- Automatic title generation for each panel
- Consistent axes and scaling
- Clean, publication-ready multi-panel figures

**Use Cases:**
- Comparing relationships across multiple categories
- Faceted analysis of data subsets
- Creating small multiples for pattern recognition
- Exploring interactions between multiple variables

---

```python
# Create faceted scatter plots to compare bill-tip relationships between smokers and non-smokers
# relplot automatically creates separate subplots for each category

# NOTE: plt.figure() is redundant with relplot (figure-level function)
# relplot creates and manages its own figure with appropriate sizing
plt.figure(figsize=(12, 8))

# Create faceted scatter plots:
# x='total_bill': bill amount on x-axis
# y='tip': tip amount on y-axis
# data=df_tips: source DataFrame
# kind='scatter': specifies scatter plot type (default)
# col='smoker': creates separate columns of subplots for each smoker category (Yes/No)
#               Automatically splits data and creates side-by-side plots
#               Each subplot shows the bill-tip relationship for one smoking status
#               Makes it easy to compare tipping patterns between smokers and non-smokers
sns.relplot(x='total_bill', y='tip', data=df_tips, kind='scatter', col='smoker')

# Display the faceted scatter plots
plt.show()
```

```python
# Create vertically stacked scatter plots to compare bill-tip relationships by smoking status
# Using 'row' parameter creates subplots arranged vertically instead of horizontally

# NOTE: plt.figure() is not needed with relplot (figure-level function)
# relplot automatically handles figure creation and sizing
plt.figure(figsize=(14, 8))

# Create vertically faceted scatter plots:
# x='total_bill': bill amount on x-axis
# y='tip': tip amount on y-axis
# data=df_tips: source DataFrame
# kind='scatter': specifies scatter plot type
# row='smoker': creates separate ROWS of subplots for each smoker category (Yes/No)
#               Stacks plots vertically instead of side-by-side
#               Useful when you want vertical comparison
#               Each subplot maintains consistent x-axis for easier comparison
sns.relplot(x='total_bill', y='tip', data=df_tips, kind='scatter', row='smoker')

# Display the vertically stacked scatter plots
# Semicolon suppresses any extra output in Jupyter notebooks
plt.show();
```

```python
# Create a 2x2 grid of scatter plots with faceting by two categorical variables
# This creates a matrix layout showing combinations of smoker status and meal time

# NOTE: plt.figure() is redundant with relplot
# relplot automatically creates appropriately sized figures
plt.figure(figsize=(10, 6))

# Create multi-dimensional faceted scatter plots:
# x='total_bill': bill amount on x-axis
# y='tip': tip amount on y-axis
# data=df_tips: source DataFrame
# kind='scatter': specifies scatter plot type
# col='smoker': creates columns for each smoking category (Yes/No)
# row='time': creates rows for each meal time (Lunch/Dinner)
# Result: 2x2 grid showing bill-tip relationships for all combinations:
#         - Lunch + Smoker, Lunch + Non-smoker
#         - Dinner + Smoker, Dinner + Non-smoker
# Allows comparison across two dimensions simultaneously
sns.relplot(x='total_bill', y='tip', data=df_tips, kind='scatter', col='smoker', row='time')

# Display the multi-faceted grid of scatter plots
# Semicolon suppresses extra output
plt.show();
```

```python
# Advanced multi-dimensional scatter plot encoding 5 variables simultaneously
# Demonstrates Seaborn's ability to create highly informative visualizations

# NOTE: plt.figure() is unnecessary with relplot (figure-level function)
plt.figure(figsize=(10, 6))

# Create comprehensive faceted scatter plots with multiple visual encodings:
# x='total_bill': bill amount on x-axis
# y='tip': tip amount on y-axis
# data=df_tips: source DataFrame
# kind='scatter': specifies scatter plot type
# col='smoker': creates columns for smoking status (Yes/No)
# row='time': creates rows for meal time (Lunch/Dinner)
# size='size': varies point size based on party size (number of people)
#              Larger points = larger dining parties
# hue='size': color-codes points by party size
#             Adds color gradient to reinforce the size encoding
# Result: 5 dimensions visualized in one figure:
#         1. Total bill (x-axis)
#         2. Tip amount (y-axis)
#         3. Smoking status (columns)
#         4. Meal time (rows)
#         5. Party size (point size + color)
sns.relplot(x='total_bill', y='tip', data=df_tips, kind='scatter', 
            col='smoker', row='time', size='size', hue='size')

# Display the comprehensive multi-dimensional visualization
plt.show();
```

```python
# Create a scatter plot with dual visual encoding: marker style AND color for smoking status
# Adding transparency helps reveal overlapping points in dense data regions

# NOTE: plt.figure() is redundant with relplot
plt.figure(figsize=(10, 6))

# Create scatter plot with multiple visual encodings for one variable:
# x='total_bill': bill amount on x-axis
# y='tip': tip amount on y-axis
# data=df_tips: source DataFrame
# kind='scatter': specifies scatter plot type
# style='smoker': varies MARKER SHAPE by smoking status
#                 Different symbols for smokers vs non-smokers (e.g., circles vs X's)
# hue='smoker': varies MARKER COLOR by smoking status
#               Different colors for smokers vs non-smokers
#               Encoding the same variable twice (style + hue) reinforces distinction
# alpha=0.4: sets transparency level (0=invisible, 1=opaque)
#            Makes overlapping points visible, reveals data density
#            Helps identify clusters and patterns in crowded regions
sns.relplot(x='total_bill', y='tip', data=df_tips, kind='scatter', style='smoker', hue='smoker', alpha=0.4); 

# Add descriptive title
plt.title('Relation between total bill and tips with smoker', fontsize=14)

# Label x-axis
plt.xlabel('Total bill', fontsize=14)

# Label y-axis
plt.ylabel('Tip', fontsize=14)

# Display the scatter plot with dual encoding and transparency
plt.show()
```

## Histogram in Seaborn

Seaborn provides enhanced histogram functionality through **histplot** with better aesthetics, more options, and integrated statistical features compared to matplotlib.

**Key Features:**
- Cleaner, more polished appearance
- Easy overlay of KDE (Kernel Density Estimate) curves
- Automatic binning with smart defaults
- Support for multiple distributions on same axes
- Flexible styling and color options

**Seaborn Histogram Functions:**
- `histplot()`: Modern, feature-rich histogram (recommended)
- `displot()`: Figure-level distribution plot with faceting support
- `distplot()`: Older function (deprecated, use histplot instead)

**Advantages over Matplotlib:**
- Better default bin selection
- Easy addition of density curves
- Statistical annotations
- Seamless categorical grouping with `hue`
- Multiple distribution comparison

**Common Parameters:**
- `bins`: Number of bins (automatic if not specified)
- `kde=True`: Overlay kernel density estimate curve
- `stat`: Type of statistic ('count', 'frequency', 'density', 'probability')
- `hue`: Group and color-code by categorical variable

---

```python
# Create an enhanced histogram with KDE curve to compare bill distributions by gender
# Combines frequency counts with smooth density estimation for better insights

# Set figure size for clear visualization
plt.figure(figsize=(12, 8))

# Create histogram with multiple enhancements:
# x='total_bill': numerical variable to plot
# data=df_tips: source DataFrame
# bins=50: divide data range into 50 intervals for detailed distribution
# kde=True: overlay Kernel Density Estimate curve
#           Smooth curve showing probability density
#           Helps identify distribution shape (normal, skewed, bimodal, etc.)
# hue='sex': separate and color-code distributions by gender (Male/Female)
#            Creates overlapping histograms for comparison
#            Automatically generates legend
#            Shows if males and females have different spending patterns
sns.histplot(x='total_bill', data=df_tips, bins=50, kde=True, hue='sex')

# Add descriptive title
plt.title('Distr. of total bill in Dataset', fontsize=14)

# Label x-axis showing the variable being analyzed
plt.xlabel('Total Bill', fontsize=14)

# Label y-axis showing frequency counts
plt.ylabel('Frequency', fontsize=14)

# Display the comparative histogram with density curves
plt.show()
```

```python
# Create a histogram with KDE to compare bill distributions between smokers and non-smokers
# Reveals if smoking status correlates with different spending patterns

# Set figure size for clear visualization
plt.figure(figsize=(12, 8))

# Create comparative histogram with density estimation:
# x='total_bill': numerical variable showing bill amounts
# data=df_tips: source DataFrame
# bins=50: divide data into 50 intervals for detailed distribution view
# kde=True: add Kernel Density Estimate curve
#           Smooth probability density overlay
#           Makes distribution shape easier to interpret
# hue='smoker': separate distributions by smoking status (Yes/No)
#               Creates overlapping histograms with different colors
#               Automatically includes legend
#               Allows comparison of spending patterns between smokers and non-smokers
#               Shows if there are differences in bill amounts based on smoking preference
sns.histplot(x='total_bill', data=df_tips, bins=50, kde=True, hue='smoker')

# Add descriptive title
plt.title('Distr. of total bill in Dataset', fontsize=14)

# Label x-axis identifying the bill variable
plt.xlabel('Total Bill', fontsize=14)

# Label y-axis showing frequency counts
plt.ylabel('Frequency', fontsize=14)

# Display the overlapping histograms comparing smokers vs non-smokers
plt.show()
```

## Line Plot - Relational Plot (relplot)

Seaborn's **line plots** using `relplot` are ideal for visualizing trends, time series, and continuous relationships. They provide enhanced styling and automatic statistical aggregation.

**Key Features:**
- Automatic aggregation of multiple observations
- Built-in confidence intervals to show uncertainty
- Easy faceting with `col` and `row` parameters
- Smooth, professional styling
- Support for multiple groupings with `hue`, `style`, and `size`

**Line Plot Functions:**
- `lineplot()`: Axes-level line plot
- `relplot(kind='line')`: Figure-level line plot with faceting support

**Automatic Aggregation:**
When multiple y-values exist for the same x-value, Seaborn automatically:
- Computes the mean (or other estimator)
- Shows confidence interval as shaded region
- Handles uncertainty visualization

**Advantages over Matplotlib:**
- Automatic statistical aggregation
- Built-in confidence intervals
- Cleaner handling of multiple series
- Better default aesthetics
- Easy categorical grouping

**Use Cases:**
- Time series visualization
- Trend analysis over continuous variables
- Comparing multiple series
- Showing uncertainty with confidence intervals

---

```python
# Load a new dataset for demonstrating line plots
# Air quality data is ideal for time series and trend visualization

# Construct the full file path to the air quality dataset
# os.getcwd(): get current working directory
# os.path.join(): safely combine directory path with 'dataset' folder and CSV filename
path = os.path.join(os.getcwd(), 'dataset', 'AirQuality.csv')

# Read the CSV file into a pandas DataFrame
# sep=';': specifies semicolon as the delimiter (instead of default comma)
#          European datasets often use semicolon separators
df_air = pd.read_csv(path, sep=';')

# Display the first 5 rows to preview the dataset structure
# Shows available columns for line plot visualizations
# Likely contains time-based measurements of air quality metrics
df_air.head()
```

```python
# Clean and preprocess the air quality dataset
# Remove null values and extract hour from time string for easier plotting

# Filter out rows where Time column has null/missing values
# ~df_air['Time'].isna(): tilde (~) negates the condition, selecting non-null rows
# This ensures we only work with complete time data
df_air = df_air[~df_air['Time'].isna()]

# Extract the hour component from time strings for analysis
# Time is likely in format "HH:MM:SS", we want just the hour (first 2 characters)
time_lst = []

# Iterate through each row in the DataFrame
for i in range(len(df_air)):
    # Extract first 2 characters from Time string (the hour portion)
    # Example: "14:30:00" becomes "14"
    time_lst.append(df_air['Time'][i][:2])
    
# Replace the original Time column with extracted hour values
# Converts time strings to just hour values (as strings)
# Makes it easier to plot trends by hour of day
df_air['Time'] = time_lst

# Display the first 5 rows to verify the time extraction worked correctly
df_air.head()
```

```python
# Create a line plot to visualize how NO2 levels change throughout the day
# Seaborn automatically aggregates multiple observations per hour and shows confidence intervals

# NOTE: plt.figure() is redundant with relplot (figure-level function)
plt.figure(figsize=(14, 8))

# Create line plot with automatic aggregation:
# x='Time': hour of day on x-axis
# y='NO2(GT)': nitrogen dioxide concentration on y-axis
# data=df_air: source DataFrame
# kind='line': specifies line plot type
# markers=True: add markers at each data point for visibility
# Seaborn automatically:
#   - Aggregates multiple NO2 readings per hour (computes mean)
#   - Calculates and displays confidence interval as shaded region
#   - Sorts data by x-axis values
#   - Shows uncertainty in the measurements
sns.relplot(x='Time', y='NO2(GT)', data=df_air, kind='line', markers=True)

# Add descriptive title explaining the relationship
plt.title('Relation between Time and NO2', fontsize=14)

# Label x-axis showing time (hour of day)
plt.xlabel('Time', fontsize=14)

# Label y-axis showing NO2 concentration levels
plt.ylabel('NO2(GT)', fontsize=14)

# Rotate x-axis labels 90 degrees for better readability
# Prevents overlapping hour labels
plt.xticks(rotation=90)

# Display the line plot showing NO2 trends throughout the day
plt.show()
```
